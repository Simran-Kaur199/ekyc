```python
import cv2
import matplotlib.pyplot as plt
```

```python
cascade_path = "haarcascade_frontalface_default.xml"
image_path = "pan1.webp"
```

```python
img = cv2.imread(image_path)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_img)
plt.show()
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap= 'gray')
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(cascade_path)
```

```python
faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)
```

```python
faces
```

```python
x=0
y=0
w=0
h=0
for (a, b, c, d) in faces:
    x=a
    y=b
    w=c
    h=d
extracted_img = rgb_img[y:y+h, x:x+w]
plt.imshow(extracted_img)
plt.show()
```

```python
# find the face with the largest area
max_area = 0
largest_face = None
for (x, y, w, h) in faces:
    area = w*h
    if area > max_area:
        max_area = area
        largest_face = (x, y, w, h)
        
if largest_face is not None:
    (x, y, w, h) = largest_face
    

# Increase dimension by 15%
new_w = int(w * 1.50)
new_h = int(h * 1.50)

# Calculate new (x, y) to keep the centre of the face same
new_x = max(0, x - int((new_w - w)/2))
new_y = max(0, y - int((new_h - h)/2))

# Extracting the enlagred image
extracted_face = rgb_img[new_y:new_y+new_h, new_x:new_x+new_w]
```

```python
plt.imshow(extracted_face)
plt.axis("off")
plt.show()
```
